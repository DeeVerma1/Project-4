```python
import pandas as pd 
import missingno as msno
import matplotlib.pyplot as plt
```

```python
travel = pd.read_csv('../data/Trips_by_Distance.csv')
travel.head(3)
```

```python
travel['Level'].value_counts()
```

```python
travel.isnull().sum()
```

```python
travel.shape
```

```python
msno.matrix(travel);
```

```python
travel.dropna(inplace=True)
```

```python
travel.shape
```

```python
travel.to_csv('../data/clean_data/travel_data_clean.csv')
```

```python
travel['Date'] = pd.to_datetime(travel['Date'])

travel.info()
```

```python
travel.set_index('Date', inplace=True)
travel
```

```python
travel =travel.groupby(by=travel.index).sum()
travel = travel.resample('W').mean()

```

```python

df_not_home = travel#.loc['2021-09':'2022-09']
plt.figure(figsize=(18,4))
plt.plot(df_not_home['Population Staying at Home'].rolling(window=1).mean())
```
